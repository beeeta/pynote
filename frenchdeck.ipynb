```python
from collections import namedtuple

Card = namedtuple('Card',['rank','suit'])

def gtf(self,other):
        switch = {'J':'11','Q':'12','K':'13','A':'14','2':'15'}
        self_rank = switch.get(self.rank, None) if switch.get(self.rank, None) is not None else self.rank
        other_rank = switch.get(other.rank, None) if switch.get(other.rank, None) is not None else other.rank
        print('lilie:{}-{},and hanmeimei:{}-{}'.format(self.rank,self.suit,other.rank,other.suit))
        if int(self_rank) == int(other_rank):
            return True if Frenchdeck.suits.index(self.suit) - Frenchdeck.suits.index(other.suit) > 0 else False
        else:
            print('rank not equal,lilei:{},hanmeimei:{}'.format(self_rank,other_rank))
            return True if int(self_rank) - int(other_rank) > 0 else False
        
def eqf(self,other):
    if self.rank == other.rank and self.suit == other.suit:
        return True
    else:
        return False

Card.__gt__ = gtf
Card.__eq__ = eqf

class Frenchdeck:
    ranks = [str(i) for i in range(2,11)] + list('JQKA')
    suits = 'diamonds clubs hearts spades'.split(' ')
    
    def __init__(self):
        self._cards = [Card(rank,suit) for rank in self.ranks for suit in self.suits]
        
    def __len__(self):
        return len(self._cards)
    
    def __getitem__(self,index):
        return self._cards[index]
    
    

    
```

```python
from random import choice

lilei_cards = Frenchdeck()
hanmeimei_cards = Frenchdeck()

rounds = 1000

li_win_count = 0
han_win_count = 0
draw_count = 0

for i in range(rounds):
    
    licard = choice(lilei_cards)
    hancard = choice(hanmeimei_cards)
    
    if  licard > hancard:
        print('lilei win')
        li_win_count += 1
    elif licard == hancard:
        print('draw ')
        draw_count += 1
    else:
        print('hanmeimei win')
        han_win_count += 1
    
    if i == rounds - 1:
        if li_win_count > han_win_count:
            print('limlei win,win count:{},lose count:{},draw count:{}'.format(li_win_count,han_win_count,draw_count))
        elif li_win_count < han_win_count:
            print('hanmeimei win,win count:{},lose count:{},draw count:{}'.format(han_win_count,li_win_count,draw_count))
        else:
            print('draw ,hanmeimei win count:{},lilei count:{},draw count:{}'.format(han_win_count,li_win_count,draw_count))
    
    
```
